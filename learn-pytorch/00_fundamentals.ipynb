## Matrix multiplication

```python
# main two rules for matrix multiplication
# 1. inner dims must match
# 2. resulting matrix has the shape of the outer dims
tensor = torch.tensor([1, 2, 3])
tensor.shape
```

```python
# element-wise multiplication
# [1*1, 2*2, 3*3] = [1, 4, 9]
tensor * tensor
```

```python
# matrix multiplication
# [1*1 + 2*2 + 3*3] = [14]
tensor @ tensor # OR tensor.matmul(tensor)
```

```python
# transposing
# torch.transpose(input, dim0, dim1) -> input is the desired tensor to transpose. dim0 and dim1 are the dims to be swapped.
# tensor.T where tensor is the desired tensor to transpose.
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]], dtype=torch.float32)

tensor_B = torch.tensor([[7, 10],
                         [8, 11], 
                         [9, 12]], dtype=torch.float32)

torch.matmul(tensor_A, tensor_B.T)
```

```python
torch.mm(tensor_A, tensor_B.T) # shorthand for tensor.matmul
```

```python
torch.manual_seed(42)
linear = torch.nn.Linear(
    in_features=2, # matches the inner dim of the input
    out_features=6 # matches the outer dim of the output
)

x = tensor_A
output = linear(x) # (3, 2) x (2, 6) => (3, 6)
print(f"{x.shape=}")
print(f"{output.shape=}")
print(output)
```

## Reshaping, stacking, squeezing, and unsqueezing

```python
# Oh boy do I need this section
x = torch.arange(1., 8.)
x, x.shape
```

```python
x_reshaped = x.reshape(1, 7) # add an extra dimension
x_reshaped, x_reshaped.shape # not guaranteed whether or not reshaped shares its data or is a new tensor altogether.
```

```python
# change view
# torch.view creates a view of the original tensor. the new tensor ALWAYS shares its data with the original tensor. if you change the original tensor, the viewed tensor will change and vice versa. like aliasing.
z = x.view(1, 7)
z, z.shape
```

```python
x.fill_(1)
z
```

```python
x_reshaped
```

```python
x = torch.arange(1., 8.)
x_stacked = torch.stack([x, x, x, x], dim=0) # since x.shape is [7], this creates a tensor w shape [4, 7]
x_stacked, x_stacked.shape
```

```python
x = torch.arange(1., 8.)
x_stacked = torch.stack([x, x, x, x], dim=1) # since x.shape is [7], this creates a tensor w shape [7, 4]
x_stacked, x_stacked.shape
```
